```python
import random
# 포켓몬 클래스 정의
class Pokemon():
    def __init__(self, name, level=5):
        self.name = name
        self.level = level
        self.hp = level * 10
# 포켓몬 이름
# 포켓몬 레벨
# 체력은 레벨에 따라 결정

# 간단한 공격: 데미지는 레벨과 1부터 3 사이의 랜덤 숫자를 곱해 결정
    def attack(self, opponent):
        damage = self.level * random.randint(1, 3)
        print(f'{self.name}이(가) {opponent.name}에게 {damage}의 데미지를 주었다!')
        opponent.hp -= damage

# 체력이 0 이하이면 쓰러진 것으로 간주
    def is_knocked_out(self):
        # if self.hp <= 0:
        #     return f'{self.name}이(가) 쓰러졌다...'
        return self.hp <= 0

# 현재 상태(이름과 체력) 출력하는 메서드
    def status(self):
        print(f'{self.name} - hp: {self.hp}')

# 포켓몬 객체 생성
pairi = Pokemon('파이리')
kobugi = Pokemon('꼬부기')

# 배틀 시뮬레이션 시작
print('start!')
while True:
    # if pairi.is_knocked_out():
    #     print(f'{kobugi.name} 승리!')
    #     break
    # if kobugi.is_knocked_out():
    #     print(f'{pairi.name} 승리!')
    #     break

# 꼬부기가 파이리를 공격
    kobugi.attack(pairi)
    if pairi.is_knocked_out():
        print(f'{pairi.name}이(가) 쓰러졌다... {kobugi.name} 승리!')
        break

# 파이리가 꼬부기를 공격
    pairi.attack(kobugi)
    if kobugi.is_knocked_out():
        print(f'{kobugi.name}이(가) 쓰러졌다... {pairi.name} 승리!')
        break

# 각 포켓몬의 현재 체력을 출력
    kobugi.status()
    pairi.status()
    print('---------')
```

```python
class Person():
    
    def __init__(self, name, hp=3):
        self.name = name
        self.hp = hp

    def rsp(self):
        my_rsp = random.choice(['가위', '바위', '보'])
        print(f'{self.name}이(가) {my_rsp}를 냈습니다.')
        return my_rsp

    def compare(self,opponent):   
        m = self.rsp()
        o = opponent.rsp()

        if m == o:
            print('비겼습니다.')   
            
        if m == '가위' and o == '바위':
            print(f'{opponent.name}이(가) 이겼습니다.')
            self.hp -= 1        
            
        if m == '가위' and o == '보':
            print(f'{self.name}이(가) 이겼습니다.')
            opponent.hp -= 1
            
        if m == '바위' and o == '보':
            print(f'{opponent.name}이(가) 이겼습니다.')
            self.hp -= 1
            
        if m == '바위' and o == '가위':
            print(f'{self.name}이(가) 이겼습니다.')
            opponent.hp -= 1
            
        if m == '보' and o == '가위':
            print(f'{opponent.name}이(가) 이겼습니다.')
            self.hp -= 1
            
        if m == '보' and o == '바위':
            print(f'{self.name}이(가) 이겼습니다.')
            opponent.hp -= 1

    def finish(self):
        return self.hp <= 0


cw = Person('찬우')
sm = Person('성민')

while True:
    cw.compare(sm)
    
    if cw.finish():
        print(f'{sm.name}의 최종 승리!')
        break

    if sm.finish():
        print(f'{cw.name}의 최종 승리!')
        break
    print('--------')
```
